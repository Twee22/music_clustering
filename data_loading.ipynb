```python
import pandas as pd
import numpy as np
import librosa
from os import listdir
from os.path import isfile, join
from statistics import mean, stdev
```

```python
def load_data(path = r"C:\Users\Ilan\Documents\code\music_analysis\Data\genres_original"):
    rows_list = []
    dirs = [f for f in listdir(path)]
    for d in dirs:
        new_path = path + "\\" + d
        files = [f for f in listdir(new_path)]
        for f in files:
            try:
                file_loc = path + "\\" + d + "\\" + f
                data , sr = librosa.load(file_loc)
                row = {"filename": f, "label": d, "audio_data": data, "audio_sr": sr}
                rows_list.append(row)
            except:
                pass

    df = pd.DataFrame(rows_list) 
    #print(df)
    return df 
```

```python
def extract_features(data):
    for index, row in data.iterrows():
        chroma_stft = librosa.feature.chroma_stft(row["audio_data"], sr=row["audio_sr"])
        data.at[index, 'chroma_stft_mean'] = chroma_stft.mean()
        data.at[index, 'chroma_stft_stdev'] = chroma_stft.std()
        data.at[index, 'chroma_stft_var'] = chroma_stft.var()
        rms = librosa.feature.rms(row["audio_data"])
        data.at[index, 'rms_mean'] = rms.mean()
        data.at[index, 'rms_stdev'] = rms.std()
        data.at[index, 'rms_var'] = rms.var()
        spectral_centroid = librosa.feature.spectral_centroid(row["audio_data"], sr=row["audio_sr"])
        data.at[index, 'spectral_centroid_mean'] = spectral_centroid.mean()
        data.at[index, 'spectral_centroid_stdev'] = spectral_centroid.std()
        data.at[index, 'spectral_centroid_var'] = spectral_centroid.var()
        spectral_bandwidth = librosa.feature.spectral_bandwidth(row["audio_data"], sr=row["audio_sr"])
        data.at[index, 'spectral_bandwidth_mean'] = spectral_bandwidth.mean()
        data.at[index, 'spectral_bandwidth_stdev'] = spectral_bandwidth.std()
        data.at[index, 'spectral_bandwidth_var'] = spectral_bandwidth.var()
        rolloff = librosa.feature.spectral_rolloff(row["audio_data"], sr=row["audio_sr"])
        data.at[index, 'rolloff_mean'] = rolloff.mean()
        data.at[index, 'rolloff_stdev'] = rolloff.std()
        data.at[index, 'rolloff_var'] = rolloff.var()
        zero_crossing_rate = librosa.feature.zero_crossing_rate(row["audio_data"])
        data.at[index, 'zero_crossing_rate_mean'] = zero_crossing_rate.mean()
        data.at[index, 'zero_crossing_rate_stdev'] = zero_crossing_rate.std()
        data.at[index, 'zero_crossing_rate_var'] = zero_crossing_rate.var()
        num_mfccs = 20
        mfccs = librosa.feature.mfcc(row["audio_data"], sr=row["audio_sr"], n_mfcc=num_mfccs)
        for i in range(num_mfccs):
            data.at[index, 'mfccs_' + str(i + 1) + '_mean'] = mfccs[i].mean()
            data.at[index, 'mfccs_' + str(i + 1) + '_stdev'] = mfccs[i].std()
            data.at[index, 'mfccs_' + str(i + 1) + '_var'] = mfccs[i].var()
            
    return data
```

```python
def get_usable_data(path=r"C:\Users\Ilan\Documents\code\music_analysis\Data\genres_original"):
    data = load_data(path=path)
    data = extract_features(data)
    
    return data
```
