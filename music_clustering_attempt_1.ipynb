```python
final_data_attempt_1 = final_data.copy()

kmeans = KMeans(n_clusters=10, random_state=0)
final_data_attempt_1['cluster'] = kmeans.fit_predict(final_data_attempt_1[['rolloff_mean', 'zero_crossing_rate_mean']])

centroids = kmeans.cluster_centers_
cen_x = [i[0] for i in centroids] 
cen_y = [i[1] for i in centroids]

final_data_attempt_1['cen_x'] = final_data_attempt_1.cluster.map({0:cen_x[0], 1:cen_x[1], 2:cen_x[2]})
final_data_attempt_1['cen_y'] = final_data_attempt_1.cluster.map({0:cen_y[0], 1:cen_y[1], 2:cen_y[2]})

colors = ['#e6194b', '#ffe119', '#f58231', '#46f0f0', '#bcf60c', '#008080', '#9a6324', '#800000', '#808000', '#000075']
final_data_attempt_1['c'] = final_data_attempt_1.cluster.map({0:colors[0], 1:colors[1], 2:colors[2], 3:colors[3], 4:colors[4], 5:colors[5], 6:colors[6], 7:colors[7], 8:colors[8], 9:colors[9]})

plt.scatter(final_data_attempt_1.rolloff_mean, final_data_attempt_1.zero_crossing_rate_mean, c=final_data_attempt_1.c, alpha = 0.6, s=10)
```
