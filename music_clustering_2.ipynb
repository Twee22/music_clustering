```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.feature_selection import SelectKBest, f_classif

from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
```

```python
final_data = pd.read_csv(r"C:\Users\Ilan\Documents\code\music_analysis\Data\features_3_sec.csv")
```

```python
fs = SelectKBest(f_classif, k=2)
y = final_data['label']
X = final_data.drop(labels=['label','length', 'filename'],axis=1)
fs.fit_transform(X,y)
print(np.array(X.keys()[fs.get_support()]))
```

```python
final_data_k_means = final_data.copy()

kmeans = KMeans(n_clusters=10, random_state=0)
final_data_k_means['cluster'] = kmeans.fit_predict(final_data_k_means[['chroma_stft_mean', 'mfcc1_mean']])

colors = ['#e6194b', '#ffe119', '#f58231', '#46f0f0', '#bcf60c', '#008080', '#9a6324', '#800000', '#808000', '#000075']
final_data_k_means['c'] = final_data_k_means.cluster.map({0:colors[0], 1:colors[1], 2:colors[2], 3:colors[3], 4:colors[4], 5:colors[5], 6:colors[6], 7:colors[7], 8:colors[8], 9:colors[9]})

plt.scatter(final_data_k_means.chroma_stft_mean, final_data_k_means.mfcc1_mean, c=final_data_k_means.c, alpha = 0.6, s=10)
```

```python
final_data_ward = final_data.copy()

ward = AgglomerativeClustering(n_clusters=10)
final_data_ward['cluster'] = ward.fit_predict(final_data_ward[['chroma_stft_mean', 'mfcc1_mean']])

colors = ['#e6194b', '#ffe119', '#f58231', '#46f0f0', '#bcf60c', '#008080', '#9a6324', '#800000', '#808000', '#000075']
final_data_ward['c'] = final_data_ward.cluster.map({0:colors[0], 1:colors[1], 2:colors[2], 3:colors[3], 4:colors[4], 5:colors[5], 6:colors[6], 7:colors[7], 8:colors[8], 9:colors[9]})

plt.scatter(final_data_ward.chroma_stft_mean, final_data_ward.mfcc1_mean, c=final_data_ward.c, alpha = 0.6, s=10)
```

```python
final_data_gaussian = final_data.copy()

gaussian = GaussianMixture(n_components=10)
final_data_gaussian['cluster'] = gaussian.fit_predict(final_data_gaussian[['chroma_stft_mean', 'mfcc1_mean']])

colors = ['#e6194b', '#ffe119', '#f58231', '#46f0f0', '#bcf60c', '#008080', '#9a6324', '#800000', '#808000', '#000075']
final_data_gaussian['c'] = final_data_gaussian.cluster.map({0:colors[0], 1:colors[1], 2:colors[2], 3:colors[3], 4:colors[4], 5:colors[5], 6:colors[6], 7:colors[7], 8:colors[8], 9:colors[9]})

plt.scatter(final_data_gaussian.chroma_stft_mean, final_data_gaussian.mfcc1_mean, c=final_data_gaussian.c, alpha = 0.6, s=10)
```
